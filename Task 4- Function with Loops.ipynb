#  Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        # Rule 1: Minimum recharge amount
        if amount < 50:
            print(" Recharge amount must be at least ₹50")
        
        # Rule 2: Valid plan check
        elif amount not in valid_plans:
            print(" Invalid plan selected")
        
        else:
            print(" Recharge Successful!")
            break   # Exit loop if valid
        
        retry = input("Do you want to retry? (yes/no): ").lower()
        if retry != "yes":
            print("Recharge Cancelled.")
            break

# Run Function
validate_recharge()
```

# Problem 2: Inventory Reorder Alert System

```python
def inventory_check(products):
    for product, stock in products.items():
        if stock < 15:
            print(f" Reorder Alert: {product} (Stock: {stock})")
        else:
            print(f" Stock OK: {product} (Stock: {stock})")

# Example Dictionary
products_data = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

# Run Function
inventory_check(products_data)
```

# Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Average Marks: {average}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
student_marks = [60, 45, 70, 55]

result = process_result(student_marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak=False):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    # Add 25% extra during peak hours
    if peak:
        fare += fare * 0.25

    return fare


def cab_fare_system():
    while True:
        distance = float(input("Enter distance in km: "))
        peak_input = input("Is it peak hour? (yes/no): ").lower()

        peak = True if peak_input == "yes" else False

        total_fare = calculate_fare(distance, peak)
        print(f" Total Fare: ₹{total_fare}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Service.")
            break

# Run Function
cab_fare_system()
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def attendance_check(attendance_list):
    present_days = 0

    # Count Present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print(f"Attendance Percentage: {percentage}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

status = attendance_check(attendance)
print("Eligibility Status:", status)
```

# Problem 6: Password Strength Checker

```python
def password_checker(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    # Rule 1: Minimum length
    if len(password) < 8:
        return " Weak Password (Minimum 8 characters required)"

    # Check for digit and special character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return " Weak Password (Must include at least one digit)"

    if not has_special:
        return " Weak Password (Must include at least one special character @#$)"

    return " Strong Password"

# Example
pwd = input("Enter Password: ")
print(password_checker(pwd))
```
